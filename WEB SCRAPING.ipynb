# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).


```python
url = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
url
```

```python
imdb = BeautifulSoup(url.content)
```

```python
movies_name = imdb.find_all('td',class_="titleColumn")
Movies = []
Release_year = []
count = 0
for i in movies_name:
    if count <=99:
        for j in i.find_all('a'):
            Movies.append(j.text.replace('\n',''))
        for k in i.find_all('span'):
            Release_year.append(k.text)
    count+=1
print(len(Movies),len(Release_year))
```

```python
IMDB_Rating = []
count = 0
for i in imdb.find_all('td',class_="ratingColumn imdbRating"):
    if count<=99:
        IMDB_Rating.append(i.text.replace('\n',''))
    count+=1
len(IMDB_Rating)
```

```python
# Creating Dataframe.
Top_100_Rated_Movies = pd.DataFrame({})
Top_100_Rated_Movies['Name'] = Movies
Top_100_Rated_Movies['IMDB Rating'] = IMDB_Rating
Top_100_Rated_Movies['Release Year'] = Release_year
Top_100_Rated_Movies.head(10)
```

# 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The  scraped data should include Product Name, Price, Image URL and Average Rating.

```python
# Checking the URL response.
page = requests.get('https://www.amazon.in/best-mobile-under-20000/s?k=best+mobile+under+20000')
page
```

```python
### Great the URL response is 200 it means we can scrape the data.
```

```python
def Amazon(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    Product =[]
    Images = []
    Price = []
    Rating = []
    for i in soup.find_all('span',class_="a-size-medium a-color-base a-text-normal",limit=16):
        Product.append(i.text.strip())
    for i in soup.find_all('img',class_="s-image",limit=16):
        Images.append(i.get('src'))
    for i in soup.find_all('span',class_="a-price-whole",limit=16):
        Price.append(i.text)
    for i in soup.find_all('span',class_="a-icon-alt",limit = 16):
        Rating.append(i.text)
     # Creating DataFrame
    df = pd.DataFrame({})
    df['Product Name'] = Product
    df['Price'] = Price
    df['Image_URL'] = Images
    df['Average Rating'] = Rating
    return df
    
```

```python
Amazon('https://www.amazon.in/best-mobile-under-20000/s?k=best+mobile+under+20000')
```

# 9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.


```python
def Fresher_jobs(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    Job_title = []
    for i in soup.find_all('div',class_="heading_4_5 profile"):
        Job_title.append(i.text.strip())
    Company = []
    for i in soup.find_all('a',class_="link_display_like_text"):
        Company.append(i.text.strip())
    Location = []
    for i in soup.find_all('p',id="location_names"):
        Location.append(i.text.strip())
    x = soup.find_all('div',class_="item_body")
    CTC = []
    for i in range(1,len(x),3):
        CTC.append(x[i].text.strip())
    Apply_date = []
    for i in range(2,len(x),3):
        Apply_date.append(x[i].text.strip())
    Fresher_jobs = pd.DataFrame({})
    Fresher_jobs['Job Title'] = Job_title
    Fresher_jobs['Company'] = Company
    Fresher_jobs['Location'] = Location
    Fresher_jobs['CTC'] = CTC
    Fresher_jobs['Apply By'] = Apply_date
    return Fresher_jobs
```

# 10. Write a python program to scrape house details from https://www.nobroker.in/ for any location. It should include house title, location, area, emi and price

```python
def No_Broker(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    house_title = []
    for i in soup.find_all('h2',class_="heading-6 font-semi-bold nb__1AShY"):
        house_title.append(i.text.strip())
    Location = []
    for i in soup.find_all('div',class_="nb__2CMjv"):
        Location.append(i.text.strip())
    Area = []
    for i in soup.find_all('div',class_="nb__3oNyC"):
        Area.append(i.text.strip())
    EMI = []
    for i in soup.find_all('div',id="roomType"):
        EMI.append(i.text)
    x = soup.find_all('div',class_="font-semi-bold heading-6")
    Price = []
    for i in range(2,len(x),3):
        Price.append(x[i].text.strip())
    No_Broker = pd.DataFrame({})
    No_Broker['House Title'] = house_title
    No_Broker['Location'] = Location
    No_Broker['Area'] = Area
    No_Broker['EMI'] = EMI
    No_Broker['Price'] = Price
    return No_Broker
```
