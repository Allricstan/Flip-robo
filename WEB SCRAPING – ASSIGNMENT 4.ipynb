```python
# Importing Libraries
import selenium
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains ##To select fixtures
```

```python
url  = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
time.sleep(2)

Rank = []
rank = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in rank[:30]:
    Rank.append(i.text.replace(".",""))
print(len(Rank))

Name = []
for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[2]'):
    Name.append(i.text)
print(len(Name))

Artist = []
for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[3]'):
    Artist.append(i.text)
print(len(Artist))

Uploaded = []
for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[5]'):
    Uploaded.append(i.text)
print(len(Uploaded))

Views = []
for i in driver.find_elements_by_xpath('//div[@class="mw-parser-output"]/table[3]/tbody/tr/td[4]'):
    Views.append(i.text)
print(len(Views))

df = pd.DataFrame({})
df['Rank'] = Rank
df['Name'] =  Name
df['Artist'] = Artist
df['Upload date'] = Uploaded
df['Views'] = Views
df
```

```python
driver.get('https://www.bcci.tv/')
time.sleep(2)
action = ActionChains(driver)
international = driver.find_element_by_xpath('//div[@class="navigation__link navigation__link--has-drop-down js-navigation-link"]')
fixtures = driver.find_element_by_xpath('//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]/div/ul/li/a')
action.move_to_element(international).click(fixtures).perform()
```

```python
driver.get('http://statisticstimes.com/')
time.sleep(5)
economy = driver.find_element_by_xpath('//div[@class="navbar"]/div[2]')
india = driver.find_element_by_xpath('//div[@class="navbar"]/div[2]/div/a[3]')
action = ActionChains(driver)
action.move_to_element(economy).click(india).perform()
time.sleep(2)

try:
    dismiss_ad = driver.find_element_by_xpath('//div[@id="dismiss-button"]')
    dismiss_ad.click()
except NoSuchElementException:
    print('No ad pop-up')

GDP = []
for i in driver.find_elements_by_xpath('//div[@style="float:left;background-color:#E6E6B8;width:400px;height:800px;"]/ul/li'):
    GDP.append(i.text)
```

```python
GDP = []
GDP_url = []
for i in driver.find_elements_by_xpath('//div[@style="float:left;background-color:#E6E6B8;width:400px;height:800px;"]/ul/li/a'):
    GDP.append(i.text)
    GDP_url.append(i.get_attribute('href'))
GDP_url
```

```python
Source = []
Date = []
About = []
for i in GDP_url:
    driver.get(i)
    time.sleep(5)
    try:
        source = driver.find_element_by_xpath('//table[@class="reference"]/tbody/tr/td[2]')
        Source.append(source.text)
    except NoSuchElementException:
        Source.append('-')
    try:
        date = driver.find_element_by_xpath('//table[@class="reference"]/tbody/tr[2]/td[2]')
        Date.append(date.text)
    except:
        try:
            date = driver.find_element_by_xpath('//table[@class="reference"]/tbody/tr[3]/td[2]')
            Date.append(date.text)
        except:
            try:
                date = driver.find_element_by_xpath('//table[@class="reference"]/tbody/tr[4]/td[2]')
                Date.append(date.text)
            except NoSuchElementException:
                Date.append('-')
    try:
        about = driver.find_elements_by_xpath('//div[@class="para"]/p[1]')
        for i in about:
            About.append(i.text)
    except NoSuchElementException:
        About.append("-")
print(len(Source),len(Date),len(About))
```

```python
driver.get('https://github.com/')
time.sleep(3)
```

```python
explore = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary')
trending = driver.find_element_by_xpath('//a[@data-ga-click="(Logged out) Header, go to Trending"]').get_attribute('href')

action = ActionChains(driver)
action.move_to_element(explore).perform(driver.get(trending))
```

```python
rep_title = []
for i in driver.find_elements_by_xpath('//span[@class="text-normal"]'):
    rep_title.append(i.text.replace('/',''))

rep_descrp = []
for i in driver.find_elements_by_xpath('//p[@class="col-9 color-text-secondary my-1 pr-4"]'):
    rep_descrp.append(i.text)
    
Lang = []
lang = driver.find_elements_by_xpath('//div[@class="f6 color-text-secondary mt-2"]')
for i in lang:
    try:
        temp = driver.find_element_by_xpath('//div[@class="f6 color-text-secondary mt-2"]/span/span[2]')
        Lang.append(i.text.replace('\n*',''))
    except NoSuchElementException:
        Lang.append('-')

Contributor_count = []
for i in driver.find_elements_by_xpath('//div[@class="f6 color-text-secondary mt-2"]/a[2]'):
    Contributor_count.append(i.text)
print(len(rep_title),len(rep_descrp),len(Lang),len(Contributor_count))

#creating dataframe
df = pd.DataFrame({})
df['Repository title'] = rep_title
df['Repository description'] = rep_descrp
df['Contributors count'] = Contributor_count
df['Language used'] = Lang
df
```
