Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of  Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"  and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
### Storing url and loading url on browser.
url = 'https://www.amazon.in/'
driver.get(url)
time.sleep(4)
search = driver.find_element_by_id('twotabsearchtextbox')
x = input('Please enter product')
time.sleep(2)
search.send_keys(x)
serach_button = driver.find_element_by_id('nav-search-submit-button')
serach_button.click()
```

```python
next_page = driver.find_elements_by_xpath('//ul[@class="a-pagination"]/li/a')
page_urls  = []
for i in next_page:
    page_urls.append(i.get_attribute('href'))
len(page_urls)
```

```python
products_links = []
for i in driver.find_elements_by_xpath('//span[@class="rush-component"]/a'):
    products_links.append(i.get_attribute('href'))
len(products_links)
```

```python
Title = [] ## for storing product titles
Rating = [] ## For storing Rating of product 
Price = [] ## For storing price of product
No_of_rating = []
Expected_Delivery = []
Availability = []
```

```python
for url in page_urls[:3]:
    driver.get(url)
    time.sleep(3)
    for i in products_links:
        products_links = []
        for i in driver.find_elements_by_xpath('//span[@class="rush-component"]/a'):
        products_links.append(i.get_attribute('href'))
        time.sleep(2)
        driver.get(i)
        time.sleep(3)
        try:
            title = driver.find_element_by_id('productTitle')
            Title.append(title.text)
        except:
            Title.append('-')
        try:
            price = driver.find_element_by_xpath('//span[@id="priceblock_ourprice"]')
            Price.append(price.text)
        except:
            Price.append('-')
        try:
            rating = driver.find_element_by_xpath('//span[@class="a-size-medium a-color-base"]')
            Rating.append(rating.text)
        except:
            Rating.append('-')
        try:
            no_of_rating = driver.find_element_by_id('acrCustomerReviewText')
            No_of_rating.append(no_of_rating.text)
        except:
            No_of_rating.append('-')
        try:
            delivery = driver.find_element_by_xpath('//div[@class="a-section a-spacing-mini"]/b')
            Expected_Delivery.append(delivery.text)
        except:
            Expected_Delivery.append('-')
        try:
            avail = driver.find_element_by_xpath('//div[@class="a-section a-spacing-base }"]/span')
            Availability.append(avail.text)
        except:
            Availability.append('-')
    
    

print(len(Title),len(Rating),len(Price),len(Availability),len(No_of_rating),len(Expected_Delivery))
```

```python
df = pd.DataFrame({})
df['Title'] = Title
df['Rating'] = Rating
df['Price'] = Price
df['No of Ratings'] = No_of_rating
df['Expected Delivery'] = Expected_Delivery
df['Availability'] = Availability
df.head(10)
```

```python
### For first page data.
driver.get('https://www.digit.in/search?keyword=gaming+laptop&keywordby=products')
time.sleep(3)
product_url = []
for i in driver.find_elements_by_xpath('//div[@class="searchPage"]/a'):
    product_url.append(i.get_attribute('href'))
len(product_url)
```

```python
# next_page data.
next_page = driver.find_element_by_xpath('//a[@title="Next Page"]').get_attribute('href')
count = 1
for page in next_page:
    if count<= 4:
        driver.get(next_page)
        for i in driver.find_elements_by_xpath('//div[@class="searchPage"]/a'):
            product_url.append(i.get_attribute('href'))
        time.sleep(5)
        next_page = driver.find_element_by_xpath('//a[@title="Next Page"]').get_attribute('href')
    count+=1   
```

```python
len(product_url)
```

```python
Name = []
OS = []
Display = []
Processor = []
Memory = []
Price = []
```

```python
for i in product_url:
    driver.get(i)
    try:
        name = driver.find_element_by_xpath('//div[@class="heading-wraper"]/h1')
        Name.append(name.text)
    except:
        Name.append('-')
    try:
        os = driver.find_element_by_xpath('//div[@class="specs_details "]/ul/li[1]/div/p[2]')
        OS.append(os.text)
    except:
        OS.append('-')
    try:
        disp = driver.find_element_by_xpath('//div[@class="specs_details "]/ul/li[2]/div/p[2]')
        Display.append(disp.text)
    except:
        Display.append('-')
    try:
        processor = driver.find_element_by_xpath('//div[@class="specs_details "]/ul/li[3]/div/p[2]')
        Processor.append(processor.text)
    except:
        Processor.append('-')
    try:
        memory = driver.find_element_by_xpath('//div[@class="specs_details "]/ul/li[4]/div/p[2]')
        Memory.append(memory.text)
    except:
        Memory.append('-')
    try:
        price = driver.find_element_by_xpath('//div[@class="price"]/strong')
        Price.append(price.text)
    except:
        Price.append('-')
```

```python
df = pd.DataFrame({})
df['Name'] = Name
df['Operating system'] = OS
df['Display'] = Display
df['processor'] = Processor
df['Memory'] = Memory
df['Price'] = Price
```

# 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
driver.get('https://www.youtube.com/watch?v=e-ORhEE9VVg')
```

```python
text =[]
time = []
upvote = []
x = driver.find_elements_by_xpath('//yt-formatted-string[@class="style-scope ytd-comment-renderer"]')
for i in range(1,len(x),2):
    text.append(x[i].text)

for i in driver.find_elements_by_xpath('//span[@id="vote-count-middle"]'):
    upvote.append(i.text)
    
for i in driver.find_elements_by_xpath('//yt-formatted-string[@class="published-time-text above-comment style-scope ytd-comment-renderer"]'):
    time.append(i.text)
```

```python

print(len(upvote),len(time),len(text))
```
